```python
# Some required libraries.
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from shapely.geometry import Point

from cartogrampy.dorling_cartogram import dorling_cartogram

%matplotlib inline
```

```python
listings_df = gpd.read_file('gdfListingsAllYearByZip.geojson')
```

```python
listings_df.bcode.unique()
```

```python
listings_df.plot()
```

```python
listings_BK = listings_df.loc[listings_df['bcode'] == '36047']
```

```python
listings_BK.plot()
```

```python
# Compute Dorling Cartogram
nyc_dorling = dorling_cartogram(listings_df,'listing',iterations=10,verbose=False)
```

```python
fig, ax = plt.subplots(1,2,figsize=(16,8))
listings_df.plot(column = 'listing',cmap = 'PuRd', scheme = 'Quantiles', k=4,legend=True, ax=ax[0])
nyc_dorling.plot(color='r',ax=ax[1])
ax[0].axis('equal')
ax[1].axis('equal')
```

```python
BK_dorling = dorling_cartogram(listings_BK,'listing', id_field='zip',iterations=350,verbose=False)
```

```python
fig, ax = plt.subplots(1,2,figsize=(16,10))
listings_BK.plot(column = 'listing',cmap = 'PuRd', scheme = 'Quantiles', k=4,legend=True, ax=ax[0])
BK_dorling.plot(color='r',linewidth = 0.8,edgecolor='black',ax=ax[1])
BK_dorling.apply(lambda x: ax[1].annotate(s=x['zip'], xy=x['geometry'].centroid.coords[0], \
                                          ha='center',zorder=1),axis=1)
ax[0].axis('equal')
ax[1].axis('equal')
```
